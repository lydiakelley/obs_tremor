### 1. Read in time series of half-hourly median seismic amplitudes in the two frequency bands of interest ("high", 10-25 Hz, and "low", 3-10 Hz)
### 2. For missing NaN values when there is no data, infill values via interpolation
### 3. Smooth the time series by convolving with a Gaussian filter

```python
import datetime
from obspy.clients.fdsn.client import Client
import obspy
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family']=['Arial']
client = Client('IRIS')
import pandas as pd
import matplotlib.dates as mdates
import dask
import pickle
from dask.diagnostics import ProgressBar
import glob
import scipy.ndimage
```

```python
# Here can load in the hourly medians from a pickle file
# Due to overlap, at half hour increments!

file_name = 'hys14_hourly_medians.pickle'
with open(file_name,'rb') as handle:
    hys14 = pickle.load(handle)
```

```python
%matplotlib widget

# Gut check to look at them!
fig, ax = plt.subplots(2,figsize=(10,5))
ax[0].plot([t.datetime for t in hys14['htimes']],hys14['hlow_med'])
ax[0].plot([t.datetime for t in hys14['htimes']],hys14['hhigh_med'],alpha=0.5)
ax[1].plot([t.datetime for t in hys14['ztimes']],hys14['zlow_med'])
ax[1].plot([t.datetime for t in hys14['ztimes']],hys14['zhigh_med'],alpha=0.5)
```

## Need to start off by filling in missing values! And make sure to save the indices of which ones were infilled- will need to mask these later.
The missing values currently are NaNs.

```python
def nan_helper(y):
    """Helper to handle indices and logical indices of NaNs.

    Input:
        - y, 1d numpy array with possible NaNs
    Output:
        - nans, logical indices of NaNs
        - index, a function, with signature indices= index(logical_indices),
          to convert logical indices of NaNs to 'equivalent' indices
    Example:
        >>> # linear interpolation of NaNs
        >>> nans, x= nan_helper(y)
        >>> y[nans]= np.interp(x(nans), x(~nans), y[~nans])
    """

    return np.isnan(y), lambda z: z.nonzero()[0]

def interp_nans(y):
    """
    Linearly interpolate between nan values
    Save the indices of nan values so that we can mask them later!
    """
    y = np.array(y)
    nans, x = nan_helper(y)
    y[nans]= np.interp(x(nans), x(~nans), y[~nans])
    to_mask = [i for i,e in enumerate(nans) if e==True]
    
    return y,to_mask
```

```python
# Fill missing values
hys14['hlow_med'],hys14['hlow_mask'] = interp_nans(hys14['hlow_med'])
hys14['hhigh_med'],hys14['hhigh_mask'] = interp_nans(hys14['hhigh_med'])
hys14['zlow_med'],hys14['zlow_mask'] = interp_nans(hys14['zlow_med'])
hys14['zhigh_med'],hys14['zhigh_mask'] = interp_nans(hys14['zhigh_med'])
```

```python
%matplotlib widget

# Check that everything looks correct
fig, ax = plt.subplots(2,figsize=(10,5))
ax[0].plot([t.datetime for t in hys14['htimes']],hys14['hlow_med'])
ax[0].plot([t.datetime for t in hys14['htimes']],hys14['hhigh_med'],alpha=0.5)
ax[1].plot([t.datetime for t in hys14['ztimes']],hys14['zlow_med'])
ax[1].plot([t.datetime for t in hys14['ztimes']],hys14['zhigh_med'],alpha=0.5)
```

## With NaNs filled, we can now apply a Gaussian filter

```python
# Specify standard deviation (sigma) of Gaussian filter, in hours
sigma = 8 # hours
# The radius of the Gaussian filter is ~3 x sigma
```

```python
# Translate that to samples, assuming a sample every 30 minutes
sample_rate = 2 # per hour
sigma = sigma * sample_rate
```

```python
# Filter data!
hys14['hhigh_8sigma_gauss'] = scipy.ndimage.gaussian_filter1d(hys14['hhigh_med'],sigma=sigma)
hys14['hlow_8sigma_gauss'] = scipy.ndimage.gaussian_filter1d(hys14['hlow_med'],sigma=sigma)
hys14['zhigh_8sigma_gauss'] = scipy.ndimage.gaussian_filter1d(hys14['zhigh_med'],sigma=sigma)
hys14['zlow_8sigma_gauss'] = scipy.ndimage.gaussian_filter1d(hys14['zlow_med'],sigma=sigma)
```

```python
%matplotlib widget

# Check again!
fig, ax = plt.subplots(2,figsize=(10,5))
ax[0].plot([t.datetime for t in hys14['htimes']],hys14['hlow_8sigma_gauss'])
ax[0].plot([t.datetime for t in hys14['htimes']],hys14['hhigh_8sigma_gauss'],alpha=0.5)
ax[1].plot([t.datetime for t in hys14['ztimes']],hys14['zlow_8sigma_gauss'])
ax[1].plot([t.datetime for t in hys14['ztimes']],hys14['zhigh_8sigma_gauss'],alpha=0.5)
ax[0].set_title('Gaussian filtered with 8-hr radius')
```
