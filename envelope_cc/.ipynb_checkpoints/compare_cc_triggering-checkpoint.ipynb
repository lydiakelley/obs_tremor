### Read in envelope cc results, for both horizontal channels, only keep windows with cc > 0.7

```python
cc_threshold = 0.7
```

```python
file_name = 'data/HYSB1_HYS14_HHN_cc_10s_shift.pickle'
with open(file_name,'rb') as handle:
    cc_hhn = pickle.load(handle)
    
# Keep only the windows with cross correlation above our chosen threshold
# Also toss the windows which have erroneously high cross correlation- this occurs due to data outages and the way we process the data in parallel
keep_cc = [i for i,e in enumerate(cc_hhn[1]) if (e > cc_threshold) & (e < 0.99)]
cc_hhn = [cc_hhn[0][i] for i in keep_cc]

file_name = 'data/HYSB1_HYS14_HHE_cc_10s_shift.pickle'
with open(file_name,'rb') as handle:
    cc_hhe = pickle.load(handle)
    
keep_cc = [i for i,e in enumerate(cc_hhe[1]) if (e > cc_threshold) & (e < 0.99)]
cc_hhe = [cc_hhe[0][i] for i in keep_cc]
```

# Below are all scripts to check how much overlap there is between different channels/detection methods

### How many of the detections are the same between channels?

```python
def check_overlap(te,tn):
    # See if the two trigger windows overlap
    if (te[0]>tn[0]) & (te[1]<tn[1]):
        return(1)
    elif (te[1]>tn[0]) & (te[1]<tn[1]):
        return(1)
    elif (te[0]>tn[0]) & (te[0]<tn[1]):
        return(1)
    elif (tn[0]>te[0]) & (tn[1]<te[1]):
        return(1)
    else:
        return(0)
```

HYSB1

```python
# HYSB1 - between channels

# If the HHN detection is caught by HHE:
overlap = []
min_diff = []
for te in hysb1_hhn_triggers:
    print(te[0],end='\r')
    # Get index of event closest in time
    diffs = [te[0]-tn[0] for tn in hysb1_hhe_triggers]
    min_diff.append(np.min(np.abs(diffs)))
    
    overlaps = [check_overlap(te,tn) for tn in hysb1_hhe_triggers]
    
    overlap.append(np.sum(overlaps))
    
hysb1_min_diffs = min_diff
hysb1_overlap = overlap
```

```python
print(len([o for o in overlap if o>=1]))
print('/')
print(len(hysb1_hhn_triggers))
```

Envelope cross correlation

```python
# ECC - between channels

# If the HHN detection is caught by HHE:
overlap = []
min_diff = []
for te in cc_hhn:
    print(te,end='\r')
    # Get index of event closest in time
    diffs = [te-tn for tn in cc_hhe]
    min_diff.append(np.min(np.abs(diffs)))
    
    overlaps = [check_overlap([te,te+300],[tn,tn+300]) for tn in cc_hhe]
    
    overlap.append(np.sum(overlaps))
    
ecc_min_diffs = min_diff
ecc_overlap = overlap
```

```python
print(len([o for o in overlap if o>=1]))
print('/')
print(len(cc_hhe))
```
